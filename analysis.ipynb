```python
import pandas as pd
import codecs
pd.set_option('max_colwidth', 800)
```

```python
tweets = pd.read_csv('fetched_tweets.txt', sep='<BREAK>', encoding='utf-8')
tweets.columns = ['tweet']
tweets.tweet = tweets.index.values
tweets = tweets.loc[~tweets.tweet.isnull()].reset_index(drop=True)
```

```python
_leaders = codecs.open('leaders.txt', 'r', encoding='utf-8')

leaders = []
for l in _leaders.readlines():
    if l.count(', ') == 0:
        continue
    leader = l.split(', ')[-1].strip()
    leaders.append(leader)
```

```python
for wl in leaders:
    wl = wl.encode('utf-8')

tweets['leader1'] = ''
tweets['leader2'] = ''
tweets['leader3'] = ''
tweets['leader4'] = ''

for i in range(len(tweets)):
    t = tweets['tweet'][i]
    counter = 1
    for wl in leaders:
        if wl.encode('utf-8') in t:
            if counter == 1:
                tweets['leader1'][i] = wl
            elif counter == 2:
                tweets['leader2'][i] = wl
            elif counter == 3:
                tweets['leader3'][i] = wl
            elif counter == 4:
                tweets['leader4'][i] = wl
            counter +=1
```
